```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.transform(credit)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
age= credit['AGE'].dropna()

# Distribution of age, with an overlay of a density plot
age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Clients' Ages")
```

```python
def Age_Splitting(clients):
    AGE, SEX = clients
    
    if (AGE > 16 and AGE < 30):
        return 'Young Adult'
    if AGE > 30 and AGE < 55:
        return 'Adult'
    else:
        return 'Old'

credit['person'] = credit[['AGE', 'SEX']].apply(Age_Splitting, axis=1)
```

```python
credit[:10]
```

```python
sns.factorplot('EDUCATION', data=credit, kind='count', hue='person', order=[1,2,3], 
               hue_order=['Young Adult','Adult','Old'], aspect=2)
```

```python
sns.factorplot('EDUCATION', data=credit, kind='count', hue='person', col='default payment next month', order=[1,2,3], 
               hue_order=['Young Adult','Adult','Old'], aspect=1.25, height=5)
```

```python
fig = sns.FacetGrid(credit, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
fig.set(title='Distribution of Age Grouped by Gender')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='MARRIAGE', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
fig.set(title='Distribution of Age Grouped by Gender')
fig.add_legend()

```
